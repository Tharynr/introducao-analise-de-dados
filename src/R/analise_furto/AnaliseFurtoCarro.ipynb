### Definição da análise

Como já conhecemos um pouco sobre carros, vamos fazer algumas perguntas simples e tentar respondê-las através da nossa análise.

1. Ano de fabricação que sofre mais roubos?
 
2. Qual é o dia da semana em que houve mais roubo?

3. Qual é o tipo de meio de locomoção mais roubado?

4. Qual é a marca que mais roubam?

5. Qual é a média de roubo de carro por mês?

6. Mapa de região mais roubada

Vamos lá, primeiro vou quero ver a quantidade de carros roubados em relação ao ano de fabricação, minha curiosidade é ver se isso influência. Vamos começar selecionando os atributos **CIDADE_VEICULO**, **ANO_MODELO** e **DESCR_COR_VEICULO** e depois vamos filtrar pela cidade de São José dos Campos

```r
carros_sjc <- df %>% select(CIDADE_VEICULO,
                            ANO_MODELO,
                            DESCR_COR_VEICULO) %>%
  filter(CIDADE_VEICULO==("S.JOSE DOS CAMPOS"))
```

Vamos observar como ficou

```r
head(carros_sjc, 4)
```

Show, agora vamos visualizar nossos dados, para não deixar o *fill* com a cor padrão e assim gerar cores aleatórias para as cores dos carros, vou selecionar minha paletas de cores e depois colocar no *y* a quantidade de carros roubadas usando o próprio *stats* do histograma e no eixo *x* o ano de fabricação. Para deixar bonito vou usar o tema do pacote * ggthemr*

```r
# Usando um tema bonitinho :D
ggthemr::ggthemr('flat dark')

# Definição de cores por classes
cores <- c("Amarelo" = "#f1c40f",
           "Azul" = "#2980b9",
           "Branco" = "#f7f1e3",
           "Cinza" = "#95a5a6",
           "Prata" = "#bdc3c7",
           "Preta" = "#3d3d3d",
           "Verde" = "#2ecc71",
           "Vermelho" = "#e74c3c")

# Grafico de barras
ggplot(carros_sjc, aes(x=ANO_MODELO,
                       fill=DESCR_COR_VEICULO)) +
  geom_histogram(binwidth = 0.5) +
  labs(title = 'Carros roubados em Abril',
       y = 'Quantidade de carros Roubados',
       x = 'Ano de frabricação') +
  scale_fill_manual("Cores dos veículos",values=cores) +
  scale_x_continuous(breaks = seq(1984,2018,2)) +
  scale_y_continuous(breaks = seq(0,16,2)) +
  theme(axis.text.x = element_text(angle = 90, hjust = 1))
```

Análisando o gráfico que usa o modelo `stack` nas barras, é possível ver que no mês de abril intervalo de carros que mais sofrem roubos são de 2013 até 2015.

Em busca de resposta para a nossa 2º pergunta, que é, qual dia da semana ocorreu mais roubo, eu chuto sexta-feira e vocês?

Vou usar bastante **pipe** nesse para deixar mais rápido (preguiça) digamos assim. Então, basicamente vou transformar as data em *dia, mês, ano - hora e minuto* usando o pacote **lubridate** e depois vou usar a função **wday** para transformar a data em dia de semana. Vamos lá

```r
# Alterando o tema
ggthemr::ggthemr('fresh')

# Selecionado a data do BO iniciado e transformo em dmy_hm(dia, mes, ano hora e minuto)
df %>% select(BO_INICIADO) %>%
  mutate (BO_INICIADO = dmy_hm(BO_INICIADO),
          DIA_SEMANA = wday(BO_INICIADO,
                            label=TRUE,
                            abbr=FALSE)) %>%
  ggplot(aes(x=DIA_SEMANA)) +
  geom_bar() +
  labs(title = 'Roubo de carro no de Abril',
       y = 'Quantidade de carros roubados',
       x = 'Dias da semana')
```

O dia da semana que teve mais roubo no mês de Abril foi na terça-feira

Obs.: Não é possível encontrar uma tendência com análise apenas de um mês, precisamos análisar uma série maior para extrair mais informações.

Agora, que já sei que terça-feira foi o dia da semana que mais teve roubo no mês de Abril e os carros do intervalo de 2013 até 2015, quero saber qual é o tipo de automóvel que sofreu mais roubos. Para isso, vou selecionar o atributo **DESCR_TIPO_VEICULO** e vou substituir espaço **" "**  por **N/I** usando *regex*

```r
tipo_carro <- df %>% select(DESCR_TIPO_VEICULO)%>%
  mutate(DESCR_TIPO_VEICULO =
           str_replace(DESCR_TIPO_VEICULO,"^$", "N/I")) %>%
  group_by(DESCR_TIPO_VEICULO) %>%
  summarise(quant = length(DESCR_TIPO_VEICULO))
```

```r
head(tipo_carro, 5)
```

Bom, agora que tratamos nosso dados, vamos visualizar

```r
# Voltando ao tema normal
ggthemr::ggthemr_reset()

ggplot(tipo_carro, aes(x=DESCR_TIPO_VEICULO,
             y=quant,
             fill=DESCR_TIPO_VEICULO)) +
  geom_point() +
  labs(title = 'Tipos de meio de locomoção roubados',
       y = 'Quantidade total',
       x = 'Meios de locomoção') +
  theme(axis.text.x = element_text(angle = 90),
        legend.position = "none")
```

Só achei meio estranho aqui, porque sempre pensei que automóvel era qualquer tipo de veículo movimo há algum combustível e com motor elétrico, mas, enfim, temos que o tipo _AUTOMOVEL_ foi o mais roubado.

Bom, agora que sabemos que o tipo _AUTOMOVEL_ foi o mais roubado, vamos analisar qual é o fabricantes mais roubado. Removi todos os espaços do atributo *DESCR_MARCA_VEICULO* e filtrei por fabricantes específicas.

```r
marcas <- stringr::str_split(df$DESCR_MARCA_VEICULO,
                    "[:punct:]") %>%
  purrr::flatten_chr()

## Transformando para dataframe
df_marcas <- as.data.frame(marcas)

df_marcas %>% select(marcas) %>%
  filter(marcas %in% c("FIAT",
         "HONDA",
         "PEUGEOT",
         "FORD",
         "TOYOTA",
         "YAMAHA")) %>%
  ggplot(aes(x=marcas)) +
  geom_bar() +
  labs(title = 'Fabricantes mais roubadas',
                    y = 'Quantidade de furtos',
                    x = 'Fabricantes')
```

Então, dentre as fabricantes filtradas, a mais roubada no mês de Abril é a *HONDA*

Bom, já que temos a localização de cada roubo, nada mais justo que plotarmos um mapa, vou usar o pacote *leaflet* e selecionar os atributos *LAT* e *LON*

```r
mapa <- df %>%
  select(DATA = BO_INICIADO, LAT = LATITUDE,
         LONG = LONGITUDE) %>%
  mutate(DATA = dmy_hm(DATA),
         LAT = as.double(LAT),
         LONG = as.double(LONG)) %>%
  na.omit()


map <- leaflet() %>%
  addTiles() %>%
  addProviderTiles("OpenStreetMap.BlackAndWhite") %>%
  # Marcando algum ponto no mapa 
  addMarkers(lng = -46.0030986, lat = -23.1894443) %>%
  addCircleMarkers(data = mapa,
                   lng = ~LONG, lat = ~LAT,
                   opacity = 1.5)

map
```
